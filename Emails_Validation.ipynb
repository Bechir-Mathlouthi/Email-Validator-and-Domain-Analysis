```python
import pandas as pd
import requests

def validate_email(email):
    api_key = 'ema_live_YBW4rR6JCvYVvbOkxEtyraX1jlkAxnFzUfGUaqcj'
    url = f'https://app.emailvalidation.io/api/v1/verify?apiKey={api_key}&email={email}'

    try:
        response = requests.get(url)
        data = response.json()

        if 'status' in data and data['status'] == 'valid':
            return True
        else:
            return False
    except Exception as e:
        print(f"Error occurred: {e}")
        return False

# Load the Excel file
file_path = '/content/100 Emails.xlsx'
df = pd.read_excel(file_path, sheet_name='ALL EMAILS')

# Validate each email address
df['Is_Valid'] = df['Email Address'].apply(validate_email)

# Print the DataFrame with validation results
print(df)
```
